```julia
function variance(x::Array)
    #Find the variance of a given one-dimensional array. The array should contain all real numbers, no NaNs or infinite values.
    #It will spit out an error otherwise.+
    #The function will choose to use a double pass method if the array is larger than 100000 points, and single pass if shorter.
    N=length(x)
    sum=zero(x[1]);
    vars=zero(x[1])
    if N >= 100000        
        for i in 1:N
            if isnan(x[i]) == false && isfinite(x[i]) == true
                sum=sum+x[i]
            else
                error("Bad input")
            end
        end
        m=sum/N;
        
        for i in 1:N
            vars=vars+(x[i]-m)^2
        end
        totalv=(1/(N-1))*vars
        flag="Double pass used"
    else
        for i in 1:N
            if isnan(x[i]) == false && isfinite(x[i]) == true
                sum=sum+x[i];
                vars=vars+x[i]^2
            else
                error("Bad input")
            end
        end;
        m=sum/N;
        totalv=(1/(N-1))*(vars-N*m^2)
        flag="Single pass used"
    end
    print(flag)
    return totalv
end
```
